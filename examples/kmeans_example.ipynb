```python
df = pd.read_csv('/content/drive/MyDrive/Curso aprendizaje automatico/Semana 5/Clase 1/pima-indians-diabetes.csv',header=None)
df.head()
```

```python
dataset = df.iloc[:,:].values
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler((0,1))
dataset_norm = scaler.fit_transform(dataset)
print(dataset_norm[0:5, :])
```

```python
x_train = dataset_norm[0:538,:]
x_test = dataset_norm[538: ,:]

print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.cluster import KMeans
k = 3
km = KMeans(n_clusters=k,init='k-means++')
km.fit(x_train)
```

```python
dato_prueba = x_test[2,:]
dato_prueba =dato_prueba.reshape((1,9))
print(dato_prueba)
prediccion = km.predict(dato_prueba)
print(prediccion)
```

```python
centroides = km.cluster_centers_
print(centroides)
```

```python
inercia = km.inertia_
print(inercia)
```
